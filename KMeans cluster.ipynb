## Importing the necessaruy libraries 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings("ignore")
```

## Importing the  data and split it

```python
data = pd.read_csv("Mall_Customers.csv")
x = data.iloc[:, 3:].values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    cluster = KMeans(n_clusters = i , init ="k-means++" ,random_state =42)
    cluster.fit(x)
    wcss.append(cluster.inertia_)
plt.scatter(range(1,11), wcss , color = "blue" )
plt.plot(range(1,11),wcss )
plt.xlabel("Range")
plt.xlabel("Annual Income $")
plt.show()
```

## Training KMeans model 

```python
kmeans = KMeans(n_clusters = 5 , init='k-means++' , random_state = 42)
pred = kmeans.fit_predict(x)
pred
```

## Visualizing KMeans Cluester

```python
plt.scatter(x[pred==0,0] , x[pred==0,1] , s =200 , c = 'red', label="Cluster 1")
plt.scatter(x[pred==1,0] , x[pred==1,1] , s =200 , c = 'blue', label="Cluster 2")
plt.scatter(x[pred==2,0] , x[pred==2,1] , s =200 , c = 'green', label="Cluster 3")
plt.scatter(x[pred==3,0] , x[pred==3,1] , s =200 , c = 'purple', label="Cluster 4")
plt.scatter(x[pred==4,0] , x[pred==4,1] , s =200 , c = 'yellow', label="Cluster 5")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s =200 , c = 'black', label = 'centroids')
plt.legend()
plt.title("Elbow Chat")
plt.xlabel("Annual Income ($)")
plt.ylabel("Spending Score(1-100)")
plt.show()
```
